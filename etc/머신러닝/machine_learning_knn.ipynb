# KNN

```python
import pandas as pd
import numpy as np
df = pd.read_csv('beer_data.csv')
df
```

```python
df = df[['bold', 'sparkling', 'sweet', 'tannic', 'acidic', 'alcohol']]
df = df.iloc[:-1, :]
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df.alcohol = scaler.fit_transform(pd.DataFrame(df.alcohol))
df.alcohol = round(df.alcohol, 2)
df
```

```python
# label 칼럼 생성
import random

y = []
for _ in range(121):
    y.append(random.randint(1, 5))
y
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors = 3)
train = df.iloc[:80, :]
test = df.iloc[80:, :]
y1 = pd.DataFrame(y).iloc[:80, :]
y2 = pd.DataFrame(y).iloc[80:, :]
classifier.fit(train, y1)
x_test = classifier.predict(test)
x_test
```
